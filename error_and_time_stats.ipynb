```python
import numpy as np
import matplotlib.pyplot as plt

data = np.load('data_test.npz')
t_data = data['t_data']
x_data = data['x_data']
```

```python
#本小节采集测试集上误差的情况
error_matrix = np.load('error_matrix.npy')
error_average = np.mean(error_matrix, axis=2)
plt.plot(t_data, error_average[0,:], label='x1 mean error')
plt.plot(t_data, error_average[1,:], label='x2 mean error')
plt.xlabel('time')
plt.ylabel('Mean Error')
plt.legend()
```

```python
error_matrix = np.load('error_matrix.npy')
error_std = np.std(error_matrix, axis=2)
plt.plot(t_data, error_std[0,:], label='x1 error standard deviation')
plt.plot(t_data, error_std[1,:], label='x2 error standard deviation')
plt.xlabel('time')
plt.ylabel('standard deviation')
plt.legend()
```

```python
#下面采集训练用时的信息
epochs_vector = np.load('epochs_vector_cpu.npy')
time_spent_vector = np.load('time_spent_vector_cpu.npy', allow_pickle=True)
plt.hist(epochs_vector, bins=20)
plt.xlabel('Epochs Used')
plt.ylabel('Number of Experiments')
```

```python
plt.hist(time_spent_vector,bins = 20)
plt.xlabel('Time Used/seconds')
plt.ylabel('Number of Experiments')
```
